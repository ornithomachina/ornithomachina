```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Loading dataset
df = pd.read_csv('dataset.csv', usecols=[1,2,3,4], na_values=['nan', 'NaN'], header=None, skip_blank_lines=True)
df.columns = df.iloc[0]
df = df.drop([0])
df.dropna(inplace=True)
print(df)

#Train/Test/Validation splitting will be 64% Train, 20% Test, and 16% Validation
#first need to split into test and train

# TODO: split into sets from the full dataframe, THEN slice out features and labels 
# High dimensionality, do i need multinomial? Can regular work? (wont be able to plot, probably )
N_train, N_test = train_test_split(df, test_size=0.2, random_state=1)

#then split into validation and train 
N_train, N_val = train_test_split(N_train, test_size=0.2, random_state=1)

print(N_train)
print(len(N_test))
print(len(N_val))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(strip_accents='ascii', stop_words='english')


X_train = np.array(N_train['text'])
X_test = np.array(N_test['text'])
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
X_train = X_train.toarray()
X_test = X_test.toarray()

#false = 0, real = 1
y_train = np.array(N_train['label']).astype(int)
y_test = np.array(N_test['label']).astype(int)
y_train = y_train.ravel()
y_test = y_test.ravel()

#not sure if I need this
new_df = pd.DataFrame(X_train)

print(X_train)
print(y_train)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
train_pred = model.predict(X_test)
train_proba = model.predict_proba(X_test[0].reshape(1, -1))

print(train_proba)
print(train_pred)
print(y_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, confusion_matrix

print('Accuracy score: ', accuracy_score(y_test, train_pred))
print('Precision score: ', precision_score(y_test, train_pred))
print('Recall score: ', recall_score(y_test, train_pred))
print(classification_report(y_test, train_pred))
```
